```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as st



```

```python
df = pd.read_csv("../bcancer.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.diagnosis.value_counts()
```

```python
class_p = df.diagnosis.value_counts(normalize=True)
class_p
```

```python
p_m = class_p['M']
p_b = class_p["B"]
p_m, p_b
```

```python
data1 = df[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']]
data1.head(2)
```

```python
data_m = data1[data1.diagnosis == "M"]
data_b = data1[data1.diagnosis == "B"]
```

```python
test = np.array([17.99, 10.38, 122.8, 1001.0])
test
```

```python
prob_m = st.gaussian_kde(data_m["radius_mean"]).pdf(test[0]) * st.gaussian_kde(data_m["texture_mean"]).pdf(test[1]) * st.gaussian_kde(data_m["perimeter_mean"]).pdf(test[2]) *  st.gaussian_kde(data_m["area_mean"]).pdf(test[3]) * p_m
prob_m
```

```python
prob_b = st.gaussian_kde(data_b["radius_mean"]).pdf(test[0]) * st.gaussian_kde(data_b["texture_mean"]).pdf(test[1]) * st.gaussian_kde(data_b["perimeter_mean"]).pdf(test[2]) *  st.gaussian_kde(data_b["area_mean"]).pdf(test[3]) * p_b
prob_b
```

```python
if prob_m > prob_b:
    print("The test data belongs to maligant")
else:
    print("the test data belongs to benign")
```
